```python
# X_train = X_train.drop(columns=[ "age","const"])
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
# Add a constant column for the intercept
# X_train = sm.add_constant(X_train)
# Calculate VIF for each feature
vif_data = pd.DataFrame()
vif_data["feature"] = X_train.columns

vif_data["VIF"] = [variance_inflation_factor(X_train.values, i)
                   for i in range(X_train.shape[1])]

print(vif_data)
```

```python
# Set a significance level
significance_level = 0.05
X_train = sm.add_constant(X_train)
def backward_elimination(X_train , y_train, significance_level=0.05):
    # Convert to DataFrame if not already
    if not isinstance(X_train, pd.DataFrame):
        X_train = pd.DataFrame(X_train)

    # Start with all features
    eliminated = True
    while eliminated:
        eliminated = False
        # Fit the Ordinary Least Squares (OLS) model
        model = sm.OLS(y_train, X_train).fit()
        # Find the highest p-value feature
        p_values = model.pvalues.iloc[1:]  # Exclude the intercept at index 0
        max_p_value = p_values.max()
        feature_with_max_p = p_values.idxmax()

        # If the highest p-value is greater than our significance level, drop it
        if max_p_value > significance_level:
            X_train = X_train.drop(columns=[feature_with_max_p])
            eliminated = True

    return X_train, model

# Perform backward elimination
X_optimized, final_model = backward_elimination(X_train, y_train, significance_level)

print("Selected features after backward elimination:", X_optimized.columns.to_list())
print(final_model.summary())
x_final = pd.DataFrame(X_optimized)
print(x_final.columns.to_list())
```

### Stepwise Regression using K-Fold Validation

```python
# Stepwise regression using SequentialFeatureSelector
model = LinearRegression()
cv = KFold(n_splits=10, shuffle=True, random_state=0)
sfs = SFS(model,
          k_features='best',
          forward=True,
          scoring='r2',
          cv=cv)

# Fit the selector
sfs.fit(X_train, y_train)

# Selected features
selected_features = list(sfs.k_feature_names_)
print("Selected features:", selected_features)

# Convert selected features to integers
selected_features = list(map(int, selected_features))

# Use integer indices for slicing
X_selected = X_train[:, selected_features]
print("X_selected shape:", X_selected.shape)

# Assuming `X_train` is a NumPy array and you have column names in `dataset.columns`
X_train = pd.DataFrame(X_train, columns=list(all_feature_indices))

# Use column names for selection
X_selected = X_train[list(selected_features)]
print(X_selected.columns.tolist())
```
